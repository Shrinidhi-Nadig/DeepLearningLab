```python
import numpy as np

# XOR dataset
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([
    [0],
    [1],
    [1],
    [0]
])

# Sigmoid activation
def sigmoid(x):
    return 1/(1 + np.exp(-x))

# Derivative of sigmoid
def sigmoid_derivative(x):
    return x*(1-x)

np.random.seed(0)

# Weights initialization
W1 = np.random.uniform(-1,1,(2,4))   # input → hidden
b1 = np.zeros((1,4))

W2 = np.random.uniform(-1,1,(4,1))   # hidden → output
b2 = np.zeros((1,1))

learning_rate = 0.5
epochs = 20000

for epoch in range(epochs):

    # Forward pass
    hidden_input = np.dot(X, W1) + b1
    hidden_output = sigmoid(hidden_input)

    final_input = np.dot(hidden_output, W2) + b2
    final_output = sigmoid(final_input)

    # Error
    error = y - final_output

    # Backpropagation
    d_output = error * sigmoid_derivative(final_output)

    error_hidden = np.dot(d_output, W2.T)
    d_hidden = error_hidden * sigmoid_derivative(hidden_output)

    # Update weights
    W2 += np.dot(hidden_output.T, d_output) * learning_rate
    b2 += np.sum(d_output, axis=0, keepdims=True) * learning_rate

    W1 += np.dot(X.T, d_hidden) * learning_rate
    b1 += np.sum(d_hidden, axis=0, keepdims=True) * learning_rate

    # Print loss
    if epoch % 1000 == 0:
        loss = np.mean(np.square(error))
        print(f"Epoch {epoch}, Loss: {loss}")

# Final predictions
print("\nFinal Predictions:")
print(final_output)
```

```python
import numpy as np

# XOR dataset
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

# One-hot encoded labels
y = np.array([
    [1,0],
    [0,1],
    [0,1],
    [1,0]
])

# Sigmoid activation (hidden layer)
def sigmoid(x):
    return 1/(1 + np.exp(-x))

def sigmoid_derivative(x):
    return x*(1-x)

# Softmax activation
def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
    return exp_x / np.sum(exp_x, axis=1, keepdims=True)

np.random.seed(0)

# Weights initialization
W1 = np.random.uniform(-1,1,(2,4))
b1 = np.zeros((1,4))

W2 = np.random.uniform(-1,1,(4,2))  # 2 outputs for softmax
b2 = np.zeros((1,2))

learning_rate = 0.5
epochs = 20000

for epoch in range(epochs):

    # Forward pass
    hidden_input = np.dot(X, W1) + b1
    hidden_output = sigmoid(hidden_input)

    final_input = np.dot(hidden_output, W2) + b2
    final_output = softmax(final_input)

    # Error
    error = y - final_output

    # Softmax gradient (simplified with MSE)
    d_output = error

    # Backpropagation
    error_hidden = np.dot(d_output, W2.T)
    d_hidden = error_hidden * sigmoid_derivative(hidden_output)

    # Update weights
    W2 += np.dot(hidden_output.T, d_output) * learning_rate
    b2 += np.sum(d_output, axis=0, keepdims=True) * learning_rate

    W1 += np.dot(X.T, d_hidden) * learning_rate
    b1 += np.sum(d_hidden, axis=0, keepdims=True) * learning_rate

    if epoch % 1000 == 0:
        loss = np.mean(np.square(error))
        print(f"Epoch {epoch}, Loss: {loss}")

print("\nSoftmax Output Probabilities:")
print(final_output)

print("\nPredicted Classes:")
print(np.argmax(final_output, axis=1))
```
